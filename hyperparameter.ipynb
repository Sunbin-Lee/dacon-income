### xgb

```python
xgb = pd.read_csv('xgb.csv', header = None)
```

```python
val = xgb[1::2]
val_top20 = val.sort_values([6])[:20]
val_top20
```

```python
train_top20 = xgb.iloc[[i-1 for i in val_top20.index]]
train_top20
```
