```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('submission/feat_engineering_0319_depth_3.csv')
df2 = pd.read_csv('submission/feat_engineering_0319_2.csv')
df3 = pd.read_csv('submission/feat_engineering_0319_depth_4_post.csv')
df4 = pd.read_csv('submission/feat_engineering_0321.csv')
```

```python
plt.hist(df1['Income'], label='depth 3', alpha=0.5, bins=50)
plt.hist(df2['Income'], label='depth 4', alpha=0.5, bins=50)
plt.hist(df3['Income'], label='depth 4 (post)', alpha=0.5, bins=50)
plt.legend()
plt.show()
```

```python
check1 = df1.loc[df1['Income']>1000]
check2 = df2.loc[df2['Income']>1000]

plt.hist(check1['Income'], label='depth 3', alpha=0.5, bins=20)
plt.hist(check2['Income'], label='depth 4', alpha=0.5, bins=20)
plt.legend()
plt.show()
```

```python
check3 = df1.loc[df1['Income']>2000]
check4 = df2.loc[df2['Income']>2000]

plt.hist(check3['Income'], label='depth 3', alpha=0.5, bins=20)
plt.hist(check4['Income'], label='depth 4', alpha=0.5, bins=20)
plt.legend()
plt.show()
```
